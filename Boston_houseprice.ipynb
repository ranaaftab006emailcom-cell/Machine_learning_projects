```python
import kagglehub
path = kagglehub.dataset_download("fedesoriano/the-boston-houseprice-data")
```

```python
import pandas as pd
import os
df = pd.read_csv(os.path.join(path,"boston.csv"))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Splitting the data and Target

```python
X = house.drop(['MEDV'], axis=1)
Y = house['MEDV']
```

```python
print(X)
print(Y)
```

Splitting the data into Training data and  Test data

```python
X_train, X_test, Y_train, Y_test =train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)

Y_pred_lr = lr.predict(X_test)

print("Linear Regression Results:")
print("R2 Score:", r2_score(Y_test, Y_pred_lr))
print("MAE:", mean_absolute_error(Y_test, Y_pred_lr))
print("MSE:", mean_squared_error(Y_test, Y_pred_lr))
print("RMSE:", np.sqrt(mean_squared_error(Y_test, Y_pred_lr)))
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train,Y_train)
```

Evaluation

Predication on training data

```python
# accuracy for predication on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R Squarred error
score_1 = metrics.r2_score(Y_train, training_data_prediction)
```

```python
# Mean Absolute error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
```

```python
print("R squared error : ", score_1)
print("Mean Absolute error : ", score_2)
```

Visulalising the actual Prices and Predicated Prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

Predication on Test Data

```python
# accuracy for predication on test data
test_data_prediction = model.predict(X_test)
```

```python
# R Squarred error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", score_1)
print("Mean Absolute error : ", score_2)
```
